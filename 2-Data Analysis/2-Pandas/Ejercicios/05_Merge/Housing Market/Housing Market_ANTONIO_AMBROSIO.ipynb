# Housing Market

### Introduction:

This time we will create our own dataset with fictional numbers to describe a house market. As we are going to create random data don't try to reason of the numbers.

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Create 3 differents Series, each of length 100, as follows: 
1. The first a random number from 1 to 4 
2. The second a random number from 1 to 3
3. The third a random number from 10,000 to 30,000

```python
serie_1 = pd.Series(np.random.randint(1, 5, 100))
serie_2 = pd.Series(np.random.randint(1, 4, 100))
serie_3 = pd.Series(np.random.randint(10000, 30001, 100))
```

### Step 3. Let's create a DataFrame by joinning the Series by column

```python
df_series = pd.DataFrame([serie_1, serie_2, serie_3]).T
```

```python
df_series
```

### Step 4. Change the name of the columns to bedrs, bathrs, price_sqr_meter

```python
df_series.columns = ['bedrs', 'bathrs', 'price_sqr_meter']
```

```python
df_series
```

### Step 5. Create a column DataFrame with the values of the 3 Series and assign it to 'bigcolumn'

```python
pd.DataFrame(pd.concat([serie_1, serie_2, serie_3]).reset_index(drop=True), columns = ['bigcolumn'])
```
