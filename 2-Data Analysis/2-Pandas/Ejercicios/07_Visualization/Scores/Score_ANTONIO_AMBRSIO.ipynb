### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

```

### Step 2. Create the DataFrame that should look like the one below.

```python
my_data = {'first_name': ['Fito', 'Chumi', 'Beni', 'Manu', 'Luisa'], 
            'last_name': ['Cerrato', 'Graciani', 'Rios', 'Garcia', 'Gonzalez'], 
            'female': [0, 1, 1, 0, 1],
            'age': [48, 25, 36, 53, 18], 
            'preTestScore': [6, 26, 32, 3, 2],
            'postTestScore': [29, 89, 65, 61, 77]}

score = pd.DataFrame(my_data, columns =['first_name', 'last_name', 'age', 'female', 'preTestScore', 'postTestScore'])

score
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
plt.scatter(score.preTestScore, score.postTestScore, s=score.age)


plt.title("preTestScore & postTestScore")
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')

plt.show()
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
plt.scatter(score.preTestScore, 
            score.postTestScore, 
            s = score.postTestScore * 4.5, 
            c = np.arange(len(score.preTestScore)))


plt.title("preTestScore x postTestScore")
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')

plt.show()
```
